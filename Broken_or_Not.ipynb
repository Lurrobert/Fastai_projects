# Creating Blocks

```python
jars = DataBlock( # init DataBlock
    blocks=(ImageBlock, CategoryBlock), # create empty space for images and categories. Like in tensorflow
    get_items=get_image_files, # function we will use to get all the images in path we will provide to ≤jars≥ later
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label, # what is label equel to? parent label or simply – folder name where file is
    item_tfms=Resize(128))
```

```python
dls = jars.dataloaders(path) # giving it path to select all images
```

```python
dls.valid.show_batch(max_n=4, nrows=1) # take a look at it
```

```python
jars = jars.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2)) # I need to augmentate data for higher accuracy
dls = jars.dataloaders(path) # loading and AT THE SAME TIME augmenting. That is how it works
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

OOooohh. That is very wrong. I definetily need to clean it.

```python
jars = jars.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = jars.dataloaders(path)
```

# Training

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
